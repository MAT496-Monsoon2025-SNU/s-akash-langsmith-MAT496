# Run Experiments in Playground

### Setup

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langsmith import Client

example_inputs = [
("What is a shader?", "GPU program that controls the rendering of graphics"),
("What is an NPC?", "Non-playable character controlled by an AI instead of a human player"),
("What programming languages are mainly used in game development?", "C++, C#, Python, JavaScript"),
]

client = Client()
dataset_name = "Game Development Dataset"

dataset = client.create_dataset(
  dataset_name=dataset_name, description="Simple questions about developing videogames",
)

inputs = [{"question": input_prompt} for input_prompt, _ in example_inputs]
outputs = [{"output": output_answer} for _, output_answer in example_inputs]

client.create_examples(
  inputs=inputs,
  outputs=outputs,
  dataset_id=dataset.id,
)
```
